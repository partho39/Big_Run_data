```python
import pandas as pd
import seaborn as sns

```

```python
data = pd.read_csv('rundata.csv')
data.shape
print("Data row :-", data.shape[0])
print("Data coloums :-", data.shape[1])
```

```python
data.dtypes
```

```python
filtered_data = data[
    (data['Event distance/length'].isin(['50km', '50mi'])) &
    (data['Year of event'] == 2020) &
    (data['Event name'].str.extract(r'\((.*?)\)')[0] == 'USA')
]

filtered_data
```

```python
filtered_data.head(10)
filtered_data.shape
```

```python
#Remove USA from Event Name

filtered_data['Event name'] =filtered_data['Event name'].str.split('(').str.get(0)
```

```python
filtered_data.head(1)
```

```python
#Clean up Athlete age
filtered_data['Athlete age category']= 2020 - filtered_data['Athlete year of birth']
```

```python
# Remove h in Athlete performance
filtered_data['Athlete performance'] = filtered_data['Athlete performance'].str.split(' ').str.get(0)
```
